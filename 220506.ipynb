숫자 데이터를 쉽게 다루기 위한 라이브러리

```python
# 리스트를 이용해 작성한 코드
import matplotlib.pyplot as plt

t = []
p2 = []
p3 = []

for i in range(0,50,2):
  t.append(i/10)
  p2.append((i/10)**2)
  p3.append((i/10)**3)

plt.plot(t,t,'r--',t,p2,'bs',t,p3,'g^')
# 그래프 저장함수 plt.savefig('파일명')
plt.show()
```

```python
# numpy를 이용해 작성한 코드

import matplotlib.pyplot as plt
import numpy as np

t  = np.arange(0.,5.,0.2)
plt.plot(t,t,'r--', t,t**2,'bs',t,t**3,'g^')
# red dashes, blue squares and green triangles
plt.show()
```

```python
#mask 적용하기

a = np.arange(-5,5)
print(a<0)
print(a[a<0]) # true 인것만 출력

mask1 = abs(a) >3
# mask 는 참또는 거짓 담고있어 해당 값 출력 or 미출력
print(a[mask1])
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(-100, 100, 1000) # 1000개의 랜덤 값 추출
y = np.random.randint(-100, 100, 1000) # 1000개의 랜덤 값 추출
size = np.random.rand(100)*1000

mask1 = abs(x) >50 # x에 저장된 값 중 절댓값이 50보다 큰 값 걸러 냄
mask2 = abs(y) >50 # y에 저장된 값 중 절댓값이 50보다 큰 값 걸러 냄

x = x[mask1 + mask2] # mask1과 mask2 중 하나라도 만족하는 값 저장
y = y[mask1 + mask2] # mask1과 mask2 중 하나라도 만족하는 값 저장


plt.scatter(x, y, c=x, cmap = 'jet', alpha = 0.7)
plt.colorbar()
plt.show()
```

```python
# 인구 구조가 가장 비슷한 지역 출력

import csv
import matplotlib.pyplot as plt
import numpy as np

f = open('/content/drive/My Drive/age.csv','r', encoding = 'cp949')
data = csv.reader(f) 
next(data)
min=1
result=0
result_name=""
data=list(data)
name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요 : ')
for row in data : 
    if name in row[0] :
        home = np.array(row[3:], dtype =int) /int(row[2])

for row in data :     
    away = np.array(row[3:], dtype = int)/int(row[2])
    s = np.sum((home - away)**2)
    if s<min and name not in row[0]:
      min=s
      result_name=row[0]
      result=away

plt.rc('font', family='NanumBarunGothic')
plt.figure(figsize=(6,3), dpi = 150)  
plt.plot(home, label=name) 
plt.plot(result, label=result_name) 
plt.legend()
plt.show()
```
